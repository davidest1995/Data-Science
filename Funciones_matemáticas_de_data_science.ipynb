```python
# FUNCIONES LINEALES

import numpy as np
import matplotlib.pyplot as plt

N= 100
m= 10
b = 5

def f(x):
  return m*x + b

x = np.linspace(-10,10, num= N)

y= f(x)


fig, ax= plt.subplots()
ax.plot(x,y)
ax.grid()

```

```python
# FUNCIONES POLINÓMICAS


def f(x):
  return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
# FUNCIONES TRASCENDENTES

def f(x):
  return np.cos(x)

y = f(x)

plt.plot(x,y)
```

```python
def f(x):
  return np.exp(x)

y = f(x)

plt.plot(x,y)
```

```python
def f(x):
  return np.log2(x)

x = np.linspace(0.1,256, num=1000)

plt.plot(x,f(x))
```

```python
# FUNCIONES SECCIONADAS

def h(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x >= 0:
      Y[idx]=1
  return Y

N= 1000

x = np.linspace(-10,10, num= N)

plt.plot(x,h(x))
```

```python
# FUNCIÓN VALOR ABSOLUTO

def h(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x >= 0:
      Y[idx]= x 
    elif x < 0:
      Y[idx]= -x
  return Y

plt.plot(x,h(x))
plt.grid()

plt.title('Función por tramos - Valor absoluto')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-100,100,1000)
C=10

def f(x):
  return np.sin(x/C)

y=f(x) 

fig,ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')
```

```python
# FUNCIÓN SIGMOIDE

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 1/(1+ np.exp(-x))

y = f(x)

plt.plot(x,y)
plt.grid()
```
