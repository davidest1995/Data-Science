```python
izq = pd.DataFrame({'key_1': ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3']
 })
der = pd.DataFrame({'key_2': ['k0','k1','k2','k3'],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3']
 })

pd.merge(izq,der, left_on='key_1', right_on='key_2')
```

```python
izq = pd.DataFrame({'key': ['k0','k1','k2','k3'],
 'A': ['A0','A1','A2','A3'],
 'B': ['B0','B1','B2','B3']
 })

der = pd.DataFrame({'key': ['k0','k1','k2',np.nan],
 'C': ['C0','C1','C2','C3'],
 'D': ['D0','D1','D2','D3']
 })

pd.merge(izq,der,how='left')
```

```python
# JOIN INDEX MATCH

izq = pd.DataFrame({'A':['A0','A1','A2'],
                    'B':['B0','B1','B2']},
                    index=['K0','K1','K2'])

der = pd.DataFrame({'C':['C0','C1','C2'],
                    'D':['D0','D1','D2']},
                    index=['K0','K2','K3'])
#  el how puede ser 'left', 'right','outer', 'inner'
izq.join(der,how='inner')
```
